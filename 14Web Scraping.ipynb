```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.flipkart.com/search?q=mobiles%20under%2050000%20rupees&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
url
```

```python
response = requests.get(url)



response
```

```python
print(response.text)
```

```python
soup = BeautifulSoup(response.text)
soup
```

```python
phone= [soup.find_all("div",class_="RG5Slk")]
phone
```

```python
for phone in soup.find_all("div", class_="RG5Slk"):
    print(phone.text.split("(")[1].split(", ")[0])
```

```python
soup.find_all("div",class_="hZ3P6w DeU9vF")
```

```python
prices = []
for price in soup.find_all("div",class_="hZ3P6w DeU9vF"):
    print(prices.append(int(price.text[1:].replace(",",""))))
    
print(prices)  
```

```python
phone_title = []
phone_color = []


for phone in soup.find_all("div", class_="RG5Slk"):
    phone_title.append(phone.text.split("(")[0])
    phone_color.append(phone.text.split("(")[1].split(", ")[0])
  
```

```python
prices = []
for price in soup.find_all("div",class_="hZ3P6w DeU9vF"):
    prices.append(int(price.text[1:].replace(",","")))
```
